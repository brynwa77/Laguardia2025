```python
payer_code = ['MD', 'MC', 'BC']
filtered3_df = df_cleaned2[df_cleaned2['payer_code'].isin(payer_code)].sort_values(by='patient_nbr')
```

```python
filtered3_df
```

```python
filtered3_df.count()
```

```python
diagnosis_counts = filtered3_df.groupby(['payer_code', 'diag_1']).size().reset_index(name='Count')
```

```python
diagnosis_counts.count()
```

```python
diagnosis_counts.nunique()
```

```python
diagnosis_counts
```

```python
top5_per_payer = (
    diagnosis_counts
    .sort_values(['payer_code', 'Count'], ascending=[True, False])
    .groupby('payer_code')
    .head(5)
)
```

```python
top5_per_payer
```

```python
import matplotlib.pyplot as plt
```

```python
for payer, group in top5_per_payer.groupby('payer_code'):
    plt.figure()
    plt.bar(group['diag_1'], group['Count'])
    plt.title(f'Top 5 Diagnoses 1 for Payer {payer}')
    plt.xlabel('diag_1')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
# Descriptive statistics per payer code
grouped_stats = filtered3_df.groupby('payer_code')['num_medications'].describe()
print("\nStats per Payer Code:\n", grouped_stats)
```
