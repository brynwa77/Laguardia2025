```python
arrest_counts = new_df.groupby('month_year')['ARREST_DATE'].count().reset_index()
```

```python
arrest_nycboro = arrest_countsby_boro.groupby(['month_year', 'ARREST_BORO'])['ARREST_KEY'].count().reset_index()
arrest_nycboro.columns = ['month_year', 'ARREST_BORO', 'ARREST_COUNT']
```

```python
import matplotlib.pyplot as plt

pivot_df = arrest_nycboro.pivot_table(
    index='month_year',
    columns='ARREST_BORO',   # or whatever your column is named
    values='ARREST_COUNT',
    aggfunc='mean'            # or 'count', 'mean', etc., depending on what you're doing
)
```

```python
# Plotting the line graph
pivot_df.plot(marker='o', linestyle='-')
plt.title('Monthly Arrest Counts by Borough')
plt.xlabel('Month/Year')
plt.ylabel('Arrest Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='NYC Borough')
plt.show()
```

```python
arrest_ofns_desc = new_df.groupby(['month_year', 'OFNS_DESC', 'ARREST_BORO'])['ARREST_KEY'].count().reset_index()
arrest_ofns_desc.columns = ['month_year', 'offense_description', 'arrest_boro', 'arrest_count']
```

```python
arrest_ofns_desc
```

```python
top5_offenses = (
    arrest_ofns_desc.groupby('offense_description')['arrest_count']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# Step 2: Plot
plt.figure(figsize=(8, 5))
plt.bar(top5_offenses['offense_description'], top5_offenses['arrest_count'])
plt.title('Top 5 Offense Descriptions by Total Arrest Count')
plt.xlabel('Offense Description')
plt.xticks(rotation=90)
plt.ylabel('Total Arrest Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
