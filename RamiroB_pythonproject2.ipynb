```python
def map_borough(dbn):
    if pd.isna(dbn) or dbn == '': 
        return "Unknown"
    elif 'X' in dbn:
        return 'Bronx'
    elif 'K' in dbn:
        return 'Brooklyn'
    elif 'M' in dbn:
        return 'Manhattan'
    elif 'R' in dbn:
        return 'Staten Island'
    elif 'Q' in dbn:
        return 'Queens'
    else:
        return 'Unknown'
```

```python
df_cleaned['borough'] = df_cleaned['School DBN'].apply(map_borough)
```

```python
import matplotlib.pyplot as plt
```

```python
schools = ['Brooklyn High School for Law and Technology', 'Bronx High School of Science','Central Park East High School','Townsend Harris High School','Staten Island Technical High School']
filtered_df = df_highest[(df_highest['School Name'].isin(schools)) & (df_highest['Regents Exam'] == 'Algebra2/Trigonometry')]
```

```python
filtered_df
```

```python
plt.figure(figsize=(8, 5))

for school in schools:
    df_school = filtered_df[filtered_df['School Name'] == school]
    plt.plot(df_school['Year'], df_school['Mean Score'], marker='o', linestyle='-', label=school)
    
plt.xlabel('Year')
plt.ylabel('Mean Score')
plt.title('Mean Algebra2/Trigonometry Scores Over Years for Two Schools')
plt.xticks(filtered_df['Year'].unique())  # Ensure all years are visible
plt.legend()
plt.grid(True)

plt.show()
```

```python
stats_summary = filtered_df.groupby('School Name')['Mean Score'].describe()
```

```python
stats_summary
```
